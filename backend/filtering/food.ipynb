# filtering food data

```python
import pandas as pd
```

```python
food = pd.read_csv('fooddata.csv')
food.head()
```

```python
food = food[['Category']].drop_duplicates()
food.head()
```

```python
food = food.rename(columns={'Category': 'name'})
food.head()
```

```python
food = food.sort_values(by="name")
```

```python
food = food.reset_index(drop=True)
food.head()
```

```python
food = food.reset_index(drop=False)
food.head()
```

```python
food.to_csv("food.csv", index=False)
```
